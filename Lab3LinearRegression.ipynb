# Lab 3: Linear Regression

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('https://github.com/toche7/DataSets/raw/main/cost.xlsx')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
X = df[['Passenger']]
X.head()
```

```python
y = df[['Fuel Cost']]
y.head()
```

```python
plt.plot(X,y, "*")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regModel = LinearRegression()
```

```python
regModel.fit(X,y)
```

```python
regModel.score(X,y)
```

```python
regModel.coef_
```

```python
regModel.intercept_
```

```python
yPredManual =  regModel.intercept_ + regModel.coef_*X
```

```python
yPredManual
```

```python
plt.plot(X,y, "*")
plt.plot(X,yPredManual, "o")
```

```python
regModel.predict(X)
```
